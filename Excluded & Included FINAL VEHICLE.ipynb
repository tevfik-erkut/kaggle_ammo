```python
import pandas as pd
import numpy as np
df = pd.read_csv("FINAL_VEHICLE_TABLE_DATATHON-Copy1.csv")
df.drop(["Unnamed: 0"], 1, inplace = True)
df.TRAFFIC_DATE = pd.to_datetime("2021/06/01") - pd.to_datetime(df.TRAFFIC_DATE)
df.TRAFFIC_DATE = df.TRAFFIC_DATE.dt.days
df.drop(df.loc[df.TRAFFIC_DATE <93].index, inplace = True)

df = df.join(df.groupby(["BASEMODEL_CODE","TOPMODEL_CODE"])["TRAFFIC_DATE"].mean(), on = ["BASEMODEL_CODE", "TOPMODEL_CODE"], rsuffix = "_r")
df = df.join(df.groupby("TOPMODEL_CODE")["TRAFFIC_DATE"].mean(), on = "TOPMODEL_CODE", rsuffix = "_YEDEK")

df.TRAFFIC_DATE.fillna(df.TRAFFIC_DATE_r, inplace = True)
df.TRAFFIC_DATE.fillna(df.TRAFFIC_DATE_YEDEK, inplace = True)
df.drop(["TRAFFIC_DATE_YEDEK"], 1, inplace= True)
df.drop(["TRAFFIC_DATE_r"], 1, inplace= True)
```

```python
df.MOTOR_GAS_TYPE = df.MOTOR_GAS_TYPE.apply(lambda x : "Dizel" if x == "Diesel" else x)
df.loc[df.MOTOR_GAS_TYPE == "Seçiniz", "MOTOR_GAS_TYPE"] = np.nan
df.loc[df.MOTOR_GAS_TYPE == "?Benzin", "MOTOR_GAS_TYPE"] = "Benzin"
df.loc[df.MOTOR_GAS_TYPE == "Benzin/Hybrid", "MOTOR_GAS_TYPE"] = "Benzin/Hibrit"
df.loc[df.MOTOR_GAS_TYPE == "Kurşunsuz benzin", "MOTOR_GAS_TYPE"] = "Kurşunsuz Benzin"

df.loc[df.GEAR_BOX_TYPE == "manuel", "GEAR_BOX_TYPE"] = "Manuel"
df.loc[df.GEAR_BOX_TYPE == "?Handschaltung", "GEAR_BOX_TYPE"] = "Manuel"
df.loc[df.GEAR_BOX_TYPE == "Düz", "GEAR_BOX_TYPE"] = "Manuel"
df.loc[df.GEAR_BOX_TYPE == "Mekanik", "GEAR_BOX_TYPE"] = "Manuel"
df.loc[df.GEAR_BOX_TYPE == "?automatisch", "GEAR_BOX_TYPE"] = "Otomatik"
df.loc[df.GEAR_BOX_TYPE == "Otomaik", "GEAR_BOX_TYPE"] = "Otomatik"
df.loc[df.GEAR_BOX_TYPE == "Sürekli Değişken", "GEAR_BOX_TYPE"] = "Otomatik"
df.loc[df.GEAR_BOX_TYPE == "Otomatik Değişken", "GEAR_BOX_TYPE"] = "Otomatik"
df.loc[df.GEAR_BOX_TYPE == "Otomatik(DSG)", "GEAR_BOX_TYPE"] = "Otomatik (DSG)"
df.loc[df.GEAR_BOX_TYPE == "Otomatik (DSG Şanzuman)", "GEAR_BOX_TYPE"] = "Otomatik (DSG)"
df.loc[df.GEAR_BOX_TYPE == "Sürekli değişken", "GEAR_BOX_TYPE"] = "Otomatik"
```

```python
import numpy as np
f = lambda x: x.median() if np.issubdtype(x.dtype, np.number) else next(iter(x.mode()), np.nan)

df.MOTOR_GAS_TYPE.fillna(df.groupby(["BRAND_CODE", 
                                     "TOPMODEL_CODE", "BASEMODEL_CODE"])["MOTOR_GAS_TYPE"].transform(f), inplace= True)
df.MOTOR_GAS_TYPE.fillna(df.groupby(["TOPMODEL_CODE",
                                     "BASEMODEL_CODE"])["MOTOR_GAS_TYPE"].transform(f), inplace= True)
df.MOTOR_GAS_TYPE.fillna(df.groupby(["TOPMODEL_CODE"])["MOTOR_GAS_TYPE"].transform(f), inplace= True)

for i in df.BRAND_CODE.unique():
    for j in range(0,12000,1643):
        try:
            s = df.loc[(df.TRAFFIC_DATE >j) & (df.TRAFFIC_DATE < j+1643) & (df.BRAND_CODE == i), "MOTOR_GAS_TYPE"].value_counts(normalize = True)
            missing = (df.MOTOR_GAS_TYPE.isnull()) & (df.BRAND_CODE == i) & (df.TRAFFIC_DATE > j) & (df.TRAFFIC_DATE <j+1643)
            df.loc[missing,'MOTOR_GAS_TYPE'] = np.random.choice(s.index, size=len(df[missing]),p=s.values)
        except:
            continue

df.MOTOR_GAS_TYPE.fillna(df.groupby(["BRAND_CODE"])["MOTOR_GAS_TYPE"].transform(f), inplace= True)
```

```python
for i in df.BRAND_CODE.unique():
    for j in range(0,12000,1643):
        try:
            s = df.loc[(df.TRAFFIC_DATE >j) & (df.TRAFFIC_DATE < j+1643) & (df.BRAND_CODE == i), "GEAR_BOX_TYPE"].value_counts(normalize = True)
            missing = (df.GEAR_BOX_TYPE.isnull()) & (df.BRAND_CODE == i) & (df.TRAFFIC_DATE > j) & (df.TRAFFIC_DATE <j+1643)
            df.loc[missing,'GEAR_BOX_TYPE'] = np.random.choice(s.index, size=len(df[missing]),p=s.values)
        except:
            continue
df.loc[df.GEAR_BOX_TYPE.isnull(), "GEAR_BOX_TYPE"] = "Manuel"
```

```python
df.to_csv("TRAIN_ORAN_final_vehicle.csv", index = False)
```
