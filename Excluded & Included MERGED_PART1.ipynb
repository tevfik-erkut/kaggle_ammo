```python
import pandas as pd
import numpy as np
import gc 
df = pd.read_csv("TRAIN_ORAN_final_sales.csv")
df2 = pd.read_csv("TRAIN_ORAN_CUSTOMER_RELATED.csv")
df = pd.merge(df,df2, on = "BASE_CUSTOMER_ID", how = "outer")

gc.collect()
del df2
df2 = pd.read_csv("TRAIN_ORAN_sıfır_araçlar.csv")
df3 = pd.merge(df, df2, on = "VEHICLE_ID", how = "outer")
df3.drop(df3.loc[df3.BASE_CUSTOMER_ID.isnull()].index, inplace = True)
df = None
df2 = None
df3.drop_duplicates(inplace = True)

fv = pd.read_csv("TRAIN_ORAN_final_vehicle.csv")

df3 = pd.merge(df3, fv, on = "VEHICLE_ID")
df3.drop_duplicates(inplace = True)

fv = None
df = pd.read_csv("TRAIN_ORAN_servis_bilgileri.csv")

df3 = pd.merge(df3, df, on = "VEHICLE_ID", how = "left")
del df
df3.drop_duplicates(inplace = True)

df = df3
del df3
df.drop(df.loc[df.VEHICLE_ID.isnull()].index, inplace = True)
df.sıfır_araba.fillna("0", inplace = True)
df.drop_duplicates(inplace = True)
df.to_csv("TRAIN_merged_part1.csv",index = False)
```
