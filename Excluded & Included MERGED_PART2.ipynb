```python
import pandas as pd
import numpy as np
import pandas as pd
import numpy as np
f = lambda x: x.median() if np.issubdtype(x.dtype, np.number) else next(iter(x.mode()), np.nan)
df = pd.read_csv("TRAIN_merged_part1.csv")
df.head()
df.car_bought_days_before.fillna(df.cust_relation_with_car_started_days_before, inplace= True)
df.days_since_last_repair_visit_per_vehicle.fillna(df.car_bought_days_before, inplace = True)
df["BASEMODEL_CODE"] = df.groupby("BASE_CUSTOMER_ID")["BASEMODEL_CODE"].transform(lambda x: x.fillna(x.median()))
df.drop(["CUSTOMER_ID"], 1, inplace = True)

df.loc[df.Age < 30, "Age_Class"] = 0
df.loc[(df.Age >= 30) & (df.Age < 44), "Age_Class"] = 1
df.loc[(df.Age >= 43) & (df.Age < 58), "Age_Class"] = 2
df.loc[(df.Age >= 58) & (df.Age < 72), "Age_Class"] = 3
df.loc[(df.Age >= 72) & (df.Age < 86), "Age_Class"] = 4

for i in ['last_repairment_payment_per_vehicle', 'BAKIM_SAYISI', 'SERVIS_SAYISI',
       'number_of_visits_to_repair', 'total_repairment_payment_per_vehicle']:
          df[i].fillna(0, inplace = True)
        
        df.loc[(df.SF_CREATE_DATE == 0) & (df.SALESFILE_ID == 1), "SF_CREATE_DATE"] = "KEK"
        
        df.loc[(df.SF_CREATE_DATE != 0) & (df.SALESFILE_ID == 1) & (df.SF_CREATE_DATE != "KEK"), "SF_CREATE_DATE"].mean()
```

```python
df.loc[(df.SF_CREATE_DATE != 0) & (df.SALESFILE_ID == 1) & (df.SF_CREATE_DATE != "KEK"), "SF_CREATE_DATE"].mean()
```

```python
df.loc[(df.SF_CREATE_DATE == 0) & (df.SALESFILE_ID == 0), "SF_CREATE_DATE"] = 2018
df.loc[(df.SF_CREATE_DATE == "KEK") & (df.SALESFILE_ID == 1), "SF_CREATE_DATE"] = 0
df = df.join(df.loc[df.SF_CREATE_DATE >= 0].groupby("BASE_CUSTOMER_ID")["SF_CREATE_DATE"].nunique(), on = "BASE_CUSTOMER_ID", rsuffix = "_r")
```

```python
for i in df.loc[df.BRAND_CODE.notnull(),"BRAND_CODE"].unique():
    for j in df.GENDER_ID.unique():
            for a in df.Age_Class.unique():
                try:
                    s = df.loc[(df.BRAND_CODE== i) & (df.GENDER_ID == j) & (df.Age_Class == a) & (df.REQ_BRAND_CODE.notnull()) , "REQ_BRAND_CODE"].value_counts(normalize =True)
                    missing = (df.BRAND_CODE == i) & (df.GENDER_ID == j) & (df.Age_Class == a) & (df.REQ_BRAND_CODE.isnull())
                    df.loc[missing,'REQ_BRAND_CODE'] = np.random.choice(s.index, size=len(df[missing]),p=s.values)
                except:
                    continue
```

```python
for i in df.loc[df.REQ_BRAND_CODE.notnull(),"REQ_BRAND_CODE"].unique():
    for j in df.GENDER_ID.unique():
            for a in df.Age_Class.unique():
                try:
                    s = df.loc[(df.REQ_BRAND_CODE== i) & (df.GENDER_ID == j) & (df.Age_Class == a) &(df.REQ_TOPMODEL_CODE.notnull()) , "REQ_TOPMODEL_CODE"].value_counts(normalize =True)
                    missing = (df.REQ_BRAND_CODE == i) & (df.GENDER_ID == j) & (df.Age_Class == a) & (df.REQ_TOPMODEL_CODE.isnull())
                    df.loc[missing,'REQ_TOPMODEL_CODE'] = np.random.choice(s.index, size=len(df[missing]),p=s.values)
                except:
                    continue
```

```python
for i in df.loc[df.REQ_BRAND_CODE.notnull(),"REQ_BRAND_CODE"].unique():
    for j in df.GENDER_ID.unique():
            for a in df.Age_Class.unique():
                try:
                    s = df.loc[(df.REQ_BRAND_CODE== i) & (df.GENDER_ID == j) & (df.Age_Class == a) & (df.OCCUPATION.notnull()) , "OCCUPATION"].value_counts(normalize =True)
                    missing = (df.REQ_BRAND_CODE == i) & (df.GENDER_ID == j) & (df.Age_Class == a) & (df.OCCUPATION.isnull())
                    df.loc[missing,'OCCUPATION'] = np.random.choice(s.index, size=len(df[missing]),p=s.values)
                except:
                    continue
```

```python
df.drop(["req_brand_code_ratio"], 1, inplace = True)
df = df.sort_values(["BASE_CUSTOMER_ID", "SF_CREATE_DATE"])
```

```python
df = df.sort_values(["BASE_CUSTOMER_ID", "SF_CREATE_DATE"], ascending = False)
df.drop_duplicates(inplace = True)
df.STATUS.fillna(method = "ffill", inplace = True)
```

```python
df.to_csv("TRAIN_ORAN_merged_part2.csv", index = False)
```
