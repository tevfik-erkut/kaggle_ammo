```python
import pandas as pd
import numpy as np
df = pd.read_csv("MASK_SERVIS_BAKIM_DATATHON_FINAL-Copy1.csv")
df.drop(["Unnamed: 0"], 1, inplace = True)
df.CREATE_DATE = pd.to_datetime("2021/06/01")- pd.to_datetime(df.CREATE_DATE)
df.CREATE_DATE = df.CREATE_DATE.dt.days

df.CREATE_DATE = df.CREATE_DATE - 92
df.drop(df.loc[df.CREATE_DATE <0].index, inplace = True)
```

```python
df = df.join(df.groupby("VEHICLE_ID")["IS_MAINTENANCE"].sum(), on = "VEHICLE_ID", rsuffix = "_r")
df["BAKIM_SAYISI"] = df["IS_MAINTENANCE_r"]
df.drop(["IS_MAINTENANCE_r"], 1, inplace = True)
df = df.join(df.groupby("VEHICLE_ID")["IS_MAINTENANCE"].count(), on = "VEHICLE_ID", rsuffix = "_r")
df["SERVIS_SAYISI"] = df["IS_MAINTENANCE_r"] - df["BAKIM_SAYISI"]
df.drop(["IS_MAINTENANCE_r"], 1, inplace = True)
df["number_of_visits_to_repair"] = df["BAKIM_SAYISI"] + df["SERVIS_SAYISI"]
```

```python
for i in range(0, 4350,365):
    s = df.loc[(df.IS_MAINTENANCE == 1) & (df.CREATE_DATE > i) & (df.CREATE_DATE <= i+365), "TOTAL_AMOUNT_TL"].mean()
    df.loc[(df.IS_MAINTENANCE == 1) & (df.CREATE_DATE >i) & (df.CREATE_DATE <= i+365) & (df.TOTAL_AMOUNT_TL.isnull()),"TOTAL_AMOUNT_TL"] = s  

    for i in range(0, 4350,365):
    s = df.loc[(df.IS_MAINTENANCE == 0) & (df.CREATE_DATE > i) & (df.CREATE_DATE <= i+365), "TOTAL_AMOUNT_TL"].mean()
    df.loc[(df.IS_MAINTENANCE == 0) & (df.CREATE_DATE >i) & (df.CREATE_DATE <= i+365) & (df.TOTAL_AMOUNT_TL.isnull()),"TOTAL_AMOUNT_TL"] = s  
    
    
    df = df.join(df.groupby("VEHICLE_ID")["TOTAL_AMOUNT_TL"].sum(), on ="VEHICLE_ID", rsuffix= "_r")
df.rename(columns = {"TOTAL_AMOUNT_TL_r": "TOTAL_REPAIRMENT_PRICE_PER_VEHICLE"}, inplace = True)
df.drop(["IS_MAINTENANCE"], 1, inplace = True)
df.head()

df.drop_duplicates("VEHICLE_ID", keep = "last", inplace = True)
df.rename(columns= {"TOTAL_AMOUNT_TL": "last_repairment_payment_per_vehicle",
                   "TOTAL_REPAIRMENT_PRICE_PER_VEHICLE": "total_repairment_payment_per_vehicle"}, inplace = True)

df.rename(columns = {"CREATE_DATE" : "days_since_last_repair_visit_per_vehicle"}, inplace = True)

df.to_csv("TRAIN_ORAN_servis_bilgileri.csv", index =False)
```
