```python
import pandas as pd
import numpy as np
cg = pd.read_csv("FINAL_CUSTOMER_RELATED_TABLE_FOR_DATATHON-Copy1.csv")
cg.drop(["Unnamed: 0"], 1, inplace = True)
cg["START_DATE"] = pd.to_datetime("2021/06/01") - pd.to_datetime(cg.START_DATE)
cg.START_DATE = cg.START_DATE.dt.days
cg["END_DATE"] = pd.to_datetime("2021/06/01") - pd.to_datetime(cg.END_DATE)
cg.END_DATE = cg.END_DATE.dt.days
cg.START_DATE = cg.START_DATE - 92
cg.END_DATE = cg.END_DATE - 92
cg.loc[cg.END_DATE < 0, "END_DATE"] = np.nan
cg.loc[cg.END_DATE.isnull(), "END_DATE"] = 0
cg["used_day_of_cars"] = cg["START_DATE"] - cg["END_DATE"]
cg.drop_duplicates(inplace = True)
cg.FK_RELATION_STATUS_ID = cg.FK_RELATION_STATUS_ID.apply(lambda x : 0 if x == 2 else x)
cg.drop(["FK_RELATION_STATUS_EXPLANATION"], 1, inplace = True)
cg.loc[cg.END_DATE>0, "number_of_cars_owned_per_cust"] = 0
cg.loc[cg.END_DATE <= 0, "number_of_cars_owned_per_cust"] = 1

cg = cg.join(cg.groupby("BASE_CUSTOMER_ID")["number_of_cars_owned_per_cust"].sum(), on = "BASE_CUSTOMER_ID", rsuffix = "_r")
cg.drop(["number_of_cars_owned_per_cust"], 1, inplace = True)
cg.rename(columns = {"number_of_cars_owned_per_cust_r": "number_of_cars_owned_per_cust"}, inplace = True)
cg.rename(columns = {"START_DATE" : "cust_relation_with_car_started_days_before"}, inplace = True)

cg.fillna(0, inplace =True)
cg = cg.join(cg.groupby("BASE_CUSTOMER_ID")["VEHICLE_ID"].nunique(), on = "BASE_CUSTOMER_ID", rsuffix = "_r")
cg.rename(columns = {"VEHICLE_ID_r": "number_of_cars_customer_has_drived"}, inplace = True)
cg.fillna(0, inplace = True)


cg.to_csv("TRAIN_ORAN_CUSTOMER_RELATED.csv", index = False)
```
